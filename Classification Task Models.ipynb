```python
# Choose this list based on problem and response variable
classifiers = [DecisionTreeClassifier(), RandomForestClassifier(), MLPClassifier(), SVC()]
names = ["Classification Tree", "Random Forest", "Basic Neural Net Classifier", "Support Vector Classifier (default)"]
```

```python
from sklearn.datasets import make_classification
X, y = make_classification(n_samples=1000, n_features=10, n_redundant=2, n_informative=2)

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.27)
```

```python
# Train each model
from sklearn.metrics import f1_score

for name, clf in zip(names, classifiers):
        clf.fit(x_train, y_train)
        y_pred = clf.predict(x_test)
        score = f1_score(y_pred, y_test)
        score = round(score, 6)
        print("The current model specifications for the " +name+" are: ")
        print(clf)
        print("The accuracy for this model is: "+ str(score))
        print("  ")
        print("  ")
```
